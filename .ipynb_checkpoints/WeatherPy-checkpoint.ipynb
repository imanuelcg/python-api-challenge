#### Analysis:
* Its obvious that the closer the cities are to the equator, the higher temperature. 
* There is no correlation between cloudiness and the latitude of cities.
* The cities that are closer to the north pole are colder than the cities that are closer to the south pole. 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# Import API key
from api_keys import api_key


# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
#Based url for the api weather map
based_url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&"

query_url = f"{based_url}appid={api_key}&q="

#Making a list to append
city_name = []
cloudiness = []
country = []
date = []
humidity = []
lat = []
lng = []
max_temp = []
wind_speed = []
response_cities = []
counter = 0

print('Beginning Data Retrieval')
print('----------------------------')
#For loop to loop through all the list 
for x in range(len(cities)):
    #If a function work it will print the try and if not it will print the except
    try:
        post_response = requests.get(query_url + str(cities[x])).json()
        #Coord and lon because the error city has no coord and lon
        post_response['coord']['lon'] 
        city_name.append(post_response['name'])
        cloudiness.append(post_response['clouds']['all'])
        country.append(post_response['sys']['country'])
        date.append(post_response['dt'])
        humidity.append(post_response['main']['humidity'])
        lat.append(post_response['coord']['lat'])
        lng.append(post_response['coord']['lon'])
        max_temp.append(post_response['main']['temp_max'])
        wind_speed.append(post_response['wind']['speed'])     
        
        #Put a counter so if a city requests are valid, it will add the number of record.
        counter = counter + 1 
        print(f"Processing Record: {counter} | {cities[x]}")
        
    except:
        print("City not found...")
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
weathers = {'City':city_name,'Cloudiness':cloudiness,'Country':country,
            'Date':date,'Humidity':humidity,'Lat':lat,'Lng':lng,
            'Max Temp':max_temp,'Wind Speed':wind_speed}

weathers_df = pd.DataFrame(weathers)
weathers_df.head()
```

```python
#saving the file in csv folder
weathers_df.to_csv('weather.csv')
```

### Plotting the Data
* Use proper labeling of the plots using plot titles (including date of analysis) and axes labels.
* Save the plotted figures as .pngs.

#### Latitude vs. Temperature Plot

```python
plt.scatter(weathers_df['Lat'],weathers_df['Max Temp'],edgecolors='black',alpha=0.75)
plt.grid()

plt.ylabel('Max Temperature (F)')
plt.xlabel('Latitude')
plt.title('City Latitude vs. Max Temperature (10/18/2019)')

plt.tight_layout()

#save fig
plt.savefig("City Latitude vs Max Temperature")
plt.show()
```

#### Latitude vs. Humidity Plot

```python
plt.scatter(weathers_df['Lat'],weathers_df['Humidity'],edgecolors='black',alpha=0.75)
plt.grid()

plt.ylabel('Humidity (%)')
plt.xlabel('Latitude')
plt.title('City Latitude vs. Humidity (10/18/2019)')

plt.tight_layout()


#save fig
plt.savefig("City Latitude vs Humidity")
plt.show()
```

#### Latitude vs. Cloudiness Plot

```python
plt.scatter(weathers_df['Lat'],weathers_df['Cloudiness'],edgecolors='black',alpha=0.75)
plt.grid()

plt.ylabel('Cloudiness (%)')
plt.xlabel('Latitude')
plt.title('City Latitude vs. Cloudiness (10/18/2019)')
plt.tight_layout()

#save fig
plt.savefig("City latitude vs Cloudiness")
plt.show()
```

#### Latitude vs. Wind Speed Plot

```python
plt.scatter(weathers_df['Lat'],weathers_df['Wind Speed'],edgecolors='black',alpha=0.75)
plt.grid()

plt.ylabel('Wind Speed (mph)')
plt.xlabel('Latitude')
plt.title('City Latitude vs. Wind Speed (10/18/2019)')

plt.tight_layout()

#save fig
plt.savefig("City Latitude vs Wind Speed")
plt.show()
```
